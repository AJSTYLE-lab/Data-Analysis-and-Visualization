```python
for city in city_names:
    url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{city}?unitGroup=metric&elements=datetime%2Cname%2Clatitude%2Clongitude%2Ctempmax%2Ctempmin%2Ctemp%2Cfeelslikemax%2Cfeelslikemin%2Cfeelslike%2Cdew%2Chumidity%2Cprecip%2Cprecipcover%2Cwindspeed%2Cwinddir%2Cpressure%2Cvisibility%2Csolarradiation%2Csolarenergy%2Cuvindex%2Cconditions%2Cdescription%2Cicon&include=days&key=Q2J54GXWY8ED8KNVA5N5XURAR&contentType=csv"
    driver.get(url)
    text = driver.find_element_by_tag_name("pre").text.splitlines()
    csv_reader = csv.reader(text,delimiter=',', quotechar='"')
    data_list = list(csv_reader)
    df=pd.DataFrame(data_list[1:],columns=data_list[0])
    df.to_csv(f"{city}.csv")
driver.quit()
```

```python
dfs = []
for name in city_names:
    csv_file = f"{name}.csv"
    df=pd.read_csv(csv_file)
    dfs.append(df)
merged_df = pd.concat(dfs, ignore_index=True)
merged_df.to_csv('merged_data_with_names.csv', index=False)
```
